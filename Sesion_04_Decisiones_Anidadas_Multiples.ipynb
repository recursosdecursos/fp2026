# Ejercicio 2: Descuento por compra


```python
monto = float(input("Monto de compra: "))

if monto >= 200:
    descuento = monto * 0.15
elif monto >= 100:
    descuento = monto * 0.10
elif monto >= 50:
    descuento = monto * 0.05
else:
    descuento = 0

total = monto - descuento
print("Descuento:", descuento)
print("Total a pagar:", total)
```

# Ejercicio 4: Menú con `match-case`

## Problema
Mostrar una opción de menú y responder según la opción elegida.

## Idea clave
En Python, `match-case` funciona como una decisión múltiple parecida al `switch` de otros lenguajes.

## Entrada
- opción

## Proceso
- Comparar la opción usando `match-case`.

## Salida
- Mensaje según la opción elegida.


```python
print("===== Menú =====")
print("1. Saludar")
print("2. Mostrar frase")
print("3. Salir")

opcion = input("Elige una opción: ")

match opcion:
    case "1":
        print("Hola, futuro desarrollador")
    case "2":
        print("Programar es resolver problemas paso a paso")
    case "3":
        print("Programa finalizado")
    case _:
        print("Opción inválida")
```

## Nota importante sobre `match-case`

`match-case` está disponible desde Python 3.10. Si tu entorno usa una versión anterior, usa `if-elif-else` para resolver menús.
