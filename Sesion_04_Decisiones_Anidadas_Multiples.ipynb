## Mini laboratorio: orden de condiciones


```python
promedio = float(input("Promedio: "))

if promedio >= 18:
    print("Logro destacado")
elif promedio >= 14:
    print("Logro esperado")
elif promedio >= 11:
    print("En proceso")
else:
    print("En inicio")
```

# Ejercicio 1: Clasificar una nota


```python
nota = float(input("Ingrese la nota: "))

if nota < 0 or nota > 20:
    print("Nota inválida")
elif nota >= 18:
    print("Logro destacado")
elif nota >= 14:
    print("Logro esperado")
elif nota >= 11:
    print("En proceso")
else:
    print("En inicio")
```

# Ejercicio 2: Descuento por compra


```python
monto = float(input("Monto de compra: "))

if monto >= 200:
    descuento = monto * 0.15
elif monto >= 100:
    descuento = monto * 0.10
elif monto >= 50:
    descuento = monto * 0.05
else:
    descuento = 0

total = monto - descuento
print("Descuento:", descuento)
print("Total a pagar:", total)
```

# Ejercicio 3: Acceso con usuario y clave


```python
usuario = input("Usuario: ")
clave = input("Clave: ")

if usuario == "admin":
    if clave == "1234":
        print("Acceso permitido")
    else:
        print("Clave incorrecta")
else:
    print("Usuario incorrecto")
```
