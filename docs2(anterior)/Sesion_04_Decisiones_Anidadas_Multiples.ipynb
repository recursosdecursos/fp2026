# Ejercicio 1: Clasificar una nota

## Resolución guiada aplicada

1. Comprender: no basta con saber si aprueba; queremos ubicar la nota en una categoría de desempeño.
2. Entradas: `nota`.
3. Proceso: primero validar rango; luego clasificar por tramos.
4. Salida y pruebas: categoría de logro o mensaje de nota inválida. Prueba con: `19`, `15`, `11`, `8` y `25`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
nota = float(input("Ingrese la nota: "))

if nota < 0 or nota > 20:
    print("Nota inválida")
elif nota >= 18:
    print("Logro destacado")
elif nota >= 14:
    print("Logro esperado")
elif nota >= 11:
    print("En proceso")
else:
    print("En inicio")
```

# Ejercicio 2: Descuento por compra

## Resolución guiada aplicada

1. Comprender: una tienda aplica descuentos distintos según el monto comprado; el programa debe elegir el descuento correcto.
2. Entradas: `monto`.
3. Proceso: elegir porcentaje de descuento según el rango del monto.
4. Salida y pruebas: descuento y total a pagar. Prueba con: `40`, `80`, `150` y `250`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
monto = float(input("Monto de compra: "))

if monto >= 200:
    descuento = monto * 0.15
elif monto >= 100:
    descuento = monto * 0.10
elif monto >= 50:
    descuento = monto * 0.05
else:
    descuento = 0

total = monto - descuento
print("Descuento:", descuento)
print("Total a pagar:", total)
```

# Ejercicio 3: Acceso con usuario y clave

## Resolución guiada aplicada

1. Comprender: el sistema debe permitir el ingreso solo cuando usuario y clave sean correctos.
2. Entradas: `usuario` y `clave`.
3. Proceso: validar usuario; si es correcto, validar clave.
4. Salida y pruebas: acceso permitido o mensaje de error específico. Prueba con: `admin/1234`, `admin/0000` y `otro/1234`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
usuario = input("Usuario: ")
clave = input("Clave: ")

if usuario == "admin":
    if clave == "1234":
        print("Acceso permitido")
    else:
        print("Clave incorrecta")
else:
    print("Usuario incorrecto")
```

# Ejercicio 4: Menú de opciones

## Problema
Mostrar un menú y responder según la opción elegida.

## Idea clave
Un mismo problema puede tener varias respuestas posibles; para resolverlo, evaluamos cada opción con decisiones múltiples.

## Entrada
- opcion

## Proceso
- Comparar la opcion con `if-elif-else`.

## Salida
- Mensaje según la opción elegida.

## Resolución guiada aplicada

1. Comprender: un menú permite elegir una acción; el programa debe responder según la opción ingresada.
2. Entradas: `opcion` elegida por el usuario.
3. Proceso: usar `if-elif-else` para dirigir cada opción a una acción.
4. Salida y pruebas: mensaje correspondiente o Opción inválida. Prueba con: `1`, `2`, `3` y `9`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
print("===== Menú =====")
print("1. Saludar")
print("2. Mostrar frase")
print("3. Salir")

opcion = input("Elige una opción: ")

if opcion == "1":
    print("Hola, futuro desarrollador")
elif opcion == "2":
    print("Programar es resolver problemas paso a paso")
elif opcion == "3":
    print("Programa finalizado")
else:
    print("Opción inválida")
```

## Nota importante

Si todavía estás aprendiendo algoritmos, resolver menús con `if-elif-else` es suficiente y más claro para practicar decisiones múltiples.
