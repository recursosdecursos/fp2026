## Mini laboratorio: decisión simple


```python
nota = float(input("Nota: "))

if nota >= 11:
    print("Aprobó")
else:
    print("Desaprobó")
```

# Ejercicio 1: Aprobó o desaprobó

## Resolución guiada aplicada

1. Comprender: necesitamos decidir si una nota alcanza el mínimo aprobatorio de 11.
2. Entradas: `nota`.
3. Proceso: si la nota es mayor o igual que 11, aprueba; si no, desaprueba.
4. Salida y pruebas: mensaje Aprobó o Desaprobó. Prueba con: `11` debe aprobar; `10.5` debe desaprobar.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
nota = float(input("Ingrese la nota: "))

if nota >= 11:
    print("Aprobó")
else:
    print("Desaprobó")
```

# Ejercicio 2: Mayor de dos números

## Resolución guiada aplicada

1. Comprender: necesitamos comparar dos números y decir cuál es mayor, o reconocer si son iguales.
2. Entradas: `numero1` y `numero2`.
3. Proceso: comparar ambos números y considerar empate.
4. Salida y pruebas: mensaje con el mayor o aviso de igualdad. Prueba con: `8` y `5`; luego `7` y `7`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

if numero1 > numero2:
    print("El mayor es:", numero1)
elif numero2 > numero1:
    print("El mayor es:", numero2)
else:
    print("Los números son iguales")
```

# Ejercicio 3: Validar una nota

## Resolución guiada aplicada

1. Comprender: una nota escolar solo tiene sentido si está entre 0 y 20; el programa debe detectar si el dato ingresado es aceptable.
2. Entradas: `nota`.
3. Proceso: verificar si está entre 0 y 20 usando `and`.
4. Salida y pruebas: Nota válida o Nota inválida. Prueba con: `0`, `20`, `21` y `-1`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
nota = float(input("Ingrese una nota: "))

if nota >= 0 and nota <= 20:
    print("Nota válida")
else:
    print("Nota inválida")
```
