# Ejercicio 1: Calculadora básica

## 1. Problema
Leer dos números y calcular suma, resta, multiplicación y división.

## 2. Entrada
- numero1
- numero2

## 3. Proceso
- Calcular las cuatro operaciones básicas.

## 4. Salida
- Mostrar los resultados.

## 5. Código en Python

````python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

print("Suma:", numero1 + numero2)
print("Resta:", numero1 - numero2)
print("Multiplicación:", numero1 * numero2)
print("División:", numero1 / numero2)
````

# Ejercicio 2: Promedio secuencial

## 1. Problema
Calcular el promedio de tres notas.

## 2. Entrada
- nota1
- nota2
- nota3

## 3. Proceso
- promedio = (nota1 + nota2 + nota3) / 3.

## 4. Salida
- Mostrar el promedio.

## 5. Código en Python

````python
nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))
nota3 = float(input("Nota 3: "))

promedio = (nota1 + nota2 + nota3) / 3
print("Promedio:", promedio)
````

# Ejercicio 3: Área y perímetro

## 1. Problema
Calcular el área y perímetro de un rectángulo.

## 2. Entrada
- base
- altura

## 3. Proceso
- area = base * altura.
- perimetro = 2 * (base + altura).

## 4. Salida
- Mostrar área y perímetro.

## 5. Código en Python

````python
base = float(input("Base: "))
altura = float(input("Altura: "))

area = base * altura
perimetro = 2 * (base + altura)

print("Área:", area)
print("Perímetro:", perimetro)
````

# Ejercicio 4: Comparaciones sin decisiones

## 1. Problema
Comparar dos números y mostrar resultados lógicos.

## 2. Entrada
- numero1
- numero2

## 3. Proceso
- Comparar los números usando operadores relacionales.

## 4. Salida
- Mostrar `True` o `False` según cada comparación.

## 5. Código en Python

````python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

print("¿El primero es mayor?:", numero1 > numero2)
print("¿Son iguales?:", numero1 == numero2)
print("¿Son diferentes?:", numero1 != numero2)
````
