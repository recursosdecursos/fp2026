# Ejercicio 1: Promedio con clasificación

## Resolución guiada aplicada

1. Comprender: necesitamos calcular el promedio y convertirlo en una categoría comprensible para el estudiante.
2. Entradas: `nombre`, `nota1`, `nota2` y `nota3`.
3. Proceso: calcular promedio y clasificarlo con condicionales.
4. Salida y pruebas: nombre, promedio y clasificación. Prueba con: `18, 19, 20`; `11, 12, 10`; `8, 9, 10`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
nombre = input("Nombre: ")
nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))
nota3 = float(input("Nota 3: "))

promedio = (nota1 + nota2 + nota3) / 3

if promedio >= 18:
    clasificacion = "Logro destacado"
elif promedio >= 14:
    clasificacion = "Logro esperado"
elif promedio >= 11:
    clasificacion = "En proceso"
else:
    clasificacion = "En inicio"

print("Estudiante:", nombre)
print("Promedio:", promedio)
print("Clasificación:", clasificacion)
```

# Ejercicio 2: Boleta con descuento

## Resolución guiada aplicada

1. Comprender: una compra necesita subtotal, posible descuento y total final para emitir una boleta clara.
2. Entradas: `producto`, `precio` y `cantidad`.
3. Proceso: calcular subtotal, descuento y total.
4. Salida y pruebas: boleta con producto, subtotal, descuento y total. Prueba con: compra menor de 100 y compra mayor o igual a 200.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
producto = input("Producto: ")
precio = float(input("Precio: "))
cantidad = int(input("Cantidad: "))

subtotal = precio * cantidad

if subtotal >= 200:
    descuento = subtotal * 0.15
elif subtotal >= 100:
    descuento = subtotal * 0.10
else:
    descuento = 0

total = subtotal - descuento
print("Producto:", producto)
print("Subtotal:", subtotal)
print("Descuento:", descuento)
print("Total:", total)
```
