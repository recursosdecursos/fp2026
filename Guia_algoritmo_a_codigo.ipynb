# Ejercicio: Calcular el promedio de tres notas

## 1. Problema
Se necesita calcular el promedio de tres notas de un estudiante.

## 2. Entrada
- Leer nota1
- Leer nota2
- Leer nota3

## 3. Proceso
- Hacer promedio = (nota1 + nota2 + nota3) / 3.

## 4. Salida
- Mostrar "El promedio es:", promedio.

## 5. Código en Python


````python
nota1 = float(input("Ingrese la primera nota: "))
nota2 = float(input("Ingrese la segunda nota: "))
nota3 = float(input("Ingrese la tercera nota: "))

promedio = (nota1 + nota2 + nota3) / 3

print("El promedio es:", promedio)
````

# Ejercicio: Encontrar el mayor de tres números

## 1. Problema
Se necesita encontrar cuál es el número mayor entre tres números ingresados por el usuario.

## 2. Entrada
- numero1
- numero2
- numero3

## 3. Proceso
- Leer numero1, numero2 y numero3.
- Asignar a mayor el valor de numero1.
- Si numero2 es mayor que mayor, entonces mayor toma el valor de numero2.
- Si numero3 es mayor que mayor, entonces mayor toma el valor de numero3.

## 4. Salida
- Mostrar "El número mayor es:", mayor.

## 5. Código en Python

````python
# Encontrar el mayor de 3 números:
# Esta es una de las formas más fundamentales y fáciles de entender.

# Pedimos al usuario que ingrese los tres números
# Usamos int() para convertir la entrada (que es una cadena) a un número entero.
# Si quieres trabajar con números decimales, usa float() en su lugar.
numero1 = int(input("Introduce el primer número: "))
numero2 = int(input("Introduce el segundo número: "))
numero3 = int(input("Introduce el tercer número: "))

# Inicializamos una variable 'mayor' con el primer número
# Luego comparamos con los otros para encontrar el verdadero mayor
mayor = numero1

if numero2 > mayor:
    mayor = numero2

if numero3 > mayor:
    mayor = numero3

# También se podría hacer con una cadena de if/elif/else:
# if numero1 >= numero2 and numero1 >= numero3:
#     mayor = numero1
# elif numero2 >= numero1 and numero2 >= numero3:
#     mayor = numero2
# else:
#     mayor = numero3

# Imprimimos el resultado
print(f"El número mayor es: {mayor}")
````
