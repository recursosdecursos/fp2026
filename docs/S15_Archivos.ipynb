## Base conceptual de la sesión

### 1. Persistencia
#### Concepto
Es la capacidad de conservar datos más allá de la ejecución del programa mediante almacenamiento externo.

#### Elementos o clasificación
Persistencia en archivos de texto y recuperación en memoria.

#### Criterio de uso
Permite separar la memoria temporal del programa de la información que necesita guardarse para usos posteriores.

#### Ejemplo de aplicación
Guardar notas en un archivo para revisarlas después de cerrar el programa.

### 2. Archivos De Texto
#### Concepto
Son documentos simples que guardan información legible como líneas de caracteres.

#### Elementos o clasificación
Archivos línea por línea, registros separados por comas y texto plano.

#### Criterio de uso
Son adecuados para guardar datos simples y legibles. Cada línea puede representar un registro.

#### Ejemplo de aplicación
Productos.txt puede contener una línea por producto.

### 3. Escritura
#### Concepto
Es el proceso de enviar datos desde el programa hacia un archivo para conservarlos.

#### Elementos o clasificación
Escritura nueva y escritura agregada al final.

#### Criterio de uso
La escritura crea o actualiza el contenido de un archivo. Debe cuidarse si se reemplaza todo el archivo o se agrega información al final.

#### Ejemplo de aplicación
Archivo.write("Ana,18\n") guarda un registro.

### 4. Lectura
#### Concepto
Es el proceso de recuperar datos desde un archivo para mostrarlos o procesarlos dentro del programa.

#### Elementos o clasificación
Lectura completa, lectura línea por línea y separación de campos.

#### Criterio de uso
La lectura recupera datos desde un archivo y suele requerir limpieza o separación de campos antes de procesarlos.

#### Ejemplo de aplicación
For linea in archivo: print(linea.strip()).

### 5. Registro Y Recuperación
#### Concepto
Agrupa las operaciones de guardar datos y volver a leerlos posteriormente de forma coherente.

#### Elementos o clasificación
Guardar registros, leer registros y reconstruir datos.

#### Criterio de uso
Registrar y recuperar exige mantener un formato consistente para que lo guardado pueda interpretarse correctamente después.

#### Ejemplo de aplicación
Guardar "cuaderno,5.5" y luego separar la línea con split(",").

### Ejercicio 2

Guarda tres nombres de estudiantes en un archivo y luego léelos para mostrarlos numerados.

#### Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Guarda tres nombres de estudiantes en un archivo y luego léelos para mostrarlos numerados
2. Entradas: separa claramente la parte de escritura y la de lectura
3. Procesos:
- define el nombre del archivo
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- comprueba que cada línea se recupere correctamente
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
with open('estudiantes.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('Ana
')
    archivo.write('Luis
')
    archivo.write('Marta
')

with open('estudiantes.txt', 'r', encoding='utf-8') as archivo:
    for indice, linea in enumerate(archivo, start=1):
        print(indice, linea.strip())
````

### Ejercicio 3

Registra productos con su precio en un archivo de texto y luego recupéralos para mostrarlos en pantalla.

#### Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Registra productos con su precio en un archivo de texto y luego recupéralos para mostrarlos en pantalla
2. Entradas: define el formato en que guardarás cada producto
3. Procesos:
- separa escritura y lectura
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- verifica que los datos recuperados sean legibles
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
with open('productos.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('cuaderno,5.5
')
    archivo.write('lapiz,1.2
')
    archivo.write('mochila,45.0
')

with open('productos.txt', 'r', encoding='utf-8') as archivo:
    for linea in archivo:
        producto, precio = linea.strip().split(',')
        print(producto, '->', precio)
````

### Ejercicio 4

Agrega una nueva venta a un archivo existente y luego calcula el total leyendo todos los registros.

#### Resolución guiada aplicada

1. Comprender: además de crear archivos, un programa debe poder agregar datos y procesar lo guardado.
2. Entradas: ventas guardadas con formato `producto,precio`.
3. Procesos:
- usa modo `a` para agregar una línea sin borrar lo anterior
- lee el archivo completo con modo `r`
- separa cada línea con `split()` y acumula los precios convertidos a `float`
4. Salidas: registros leídos y total acumulado.
5. Pruebas: ejecuta dos veces y observa que el archivo conserva registros anteriores.
6. Código:


````python
with open('ventas.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('cuaderno,5.5\n')
    archivo.write('mochila,45.0\n')

with open('ventas.txt', 'a', encoding='utf-8') as archivo:
    archivo.write('lapicero,1.2\n')

total = 0
with open('ventas.txt', 'r', encoding='utf-8') as archivo:
    for linea in archivo:
        producto, precio = linea.strip().split(',')
        total += float(precio)
        print(producto, precio)

print('Total vendido:', round(total, 2))
````

### Tarea tipo competencia

Guarda tres productos con su precio en un archivo y luego léelos para mostrarlos en formato de lista.

#### Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código

5. Pruebas:

- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
with open('productos_persistencia.txt', 'w', encoding='utf-8') as archivo:
    archivo.write('cuaderno,5.5\n')
    archivo.write('lapiz,1.2\n')
    archivo.write('mochila,45.0\n')

with open('productos_persistencia.txt', 'r', encoding='utf-8') as archivo:
    for linea in archivo:
        producto, precio = linea.strip().split(',')
        print('- ', producto, ': ', precio, sep='')
````
