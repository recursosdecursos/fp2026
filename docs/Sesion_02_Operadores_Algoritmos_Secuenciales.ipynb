# Ejercicio 1: Calculadora básica

## Resolución guiada aplicada

1. Comprender: necesitamos una calculadora pequeña que reciba dos números y muestre varias operaciones con ellos.
2. Entradas: `numero1` y `numero2`.
3. Proceso: calcular suma, resta, multiplicación y división.
4. Salida y pruebas: mostrar cada resultado con su etiqueta. Prueba con: `10` y `2`; luego `7` y `3` para observar decimales.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

print("Suma:", numero1 + numero2)
print("Resta:", numero1 - numero2)
print("Multiplicación:", numero1 * numero2)
print("División:", numero1 / numero2)
```

# Ejercicio 2: Promedio de tres notas

## Resolución guiada aplicada

1. Comprender: necesitamos saber el rendimiento de un estudiante a partir de tres notas.
2. Entradas: `nota1`, `nota2` y `nota3`.
3. Proceso: sumar las tres notas y dividir entre 3.
4. Salida y pruebas: mostrar el promedio. Prueba con: `15, 16, 17` debe dar 16; `20, 20, 20` debe dar 20.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))
nota3 = float(input("Nota 3: "))

promedio = (nota1 + nota2 + nota3) / 3
print("El promedio es:", promedio)
```

# Ejercicio 3: Comparaciones sin decisiones

## Resolución guiada aplicada

1. Comprender: queremos observar respuestas lógicas (`True` o `False`) al comparar dos números.
2. Entradas: `numero1` y `numero2`.
3. Proceso: comparar con `>`, `==` y `!=`.
4. Salida y pruebas: mostrar valores `True` o `False`. Prueba con: `8` y `5`; luego `7` y `7`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))

print("¿El primero es mayor?:", numero1 > numero2)
print("¿Son iguales?:", numero1 == numero2)
print("¿Son diferentes?:", numero1 != numero2)
```
