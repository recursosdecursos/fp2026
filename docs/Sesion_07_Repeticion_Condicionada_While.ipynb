## Base conceptual de la sesión

### 1. While
#### Definición
repite acciones mientras una condición siga siendo verdadera.

#### Tipos principales
ciclo por condición, ciclo con centinela y ciclo de validación.

#### Ejemplo
while venta != 0: total += venta.

### 2. Condición De Parada
#### Definición
define cuándo debe detenerse el ciclo para evitar repeticiones infinitas.

#### Tipos principales
parada por valor objetivo, por centinela o por validación satisfactoria.

#### Ejemplo
terminar cuando el usuario ingresa 0.

### 3. Centinelas
#### Definición
son valores especiales usados para indicar el fin de un proceso de ingreso o repetición.

#### Tipos principales
valores de cierre como 0, fin, salir u otra marca especial.

#### Ejemplo
usar 0 como marca de cierre del ingreso.

### 4. Validación Iterativa
#### Definición
permite revisar datos dentro de un ciclo hasta que cumplan lo solicitado.

#### Tipos principales
repetición hasta que el dato cumpla una regla.

#### Ejemplo
repetir el ingreso de contraseña hasta que cumpla la regla.

### 5. Control De Ingreso
#### Definición
ayuda a mantener la calidad de los datos leídos durante la ejecución.

#### Tipos principales
lectura controlada, repetición de ingreso y verificación de regla.

#### Ejemplo
seguir pidiendo dato mientras sea inválido.

## Ejercicio 1

Registraremos ventas hasta que el usuario ingrese 0 como señal de cierre.

## Resolución guiada aplicada

1. Comprender: explica con una oración completa cuál es el resultado que debe obtener el programa.
2. Entradas: enumera cada dato de entrada e indica su tipo de dato esperado.
3. Procesos:

- ordena las acciones principales antes de escribir el código

4. Salidas: especifica la salida esperada y el formato con que se debe mostrar.

- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué sucede con una entrada vacía, inválida o fuera de rango
6. Código:


````python
total = 0
venta = float(input('Venta (0 para terminar): '))

while venta != 0:
    total += venta
    venta = float(input('Venta (0 para terminar): '))

print('Total vendido:', total)
````

## Resolución guiada aplicada

2. Entradas: define el valor centinela
3. Procesos:

````python
total = 0
numero = int(input('Número (-1 para salir): '))

while numero != -1:
    total += numero
    numero = int(input('Número (-1 para salir): '))

print('Total:', total)
````

## Tarea tipo competencia

Solicita contraseñas hasta que el usuario ingrese una que cumpla longitud mínima y contenga un número.

## Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
contrasena = input('Ingrese contraseña: ')

while True:
    tiene_numero = False
    for caracter in contrasena:
        if caracter.isdigit():
            tiene_numero = True
            break
    if len(contrasena) >= 8 and tiene_numero:
        break
    contrasena = input('Ingrese contraseña: ')

print('Contraseña válida')
````
