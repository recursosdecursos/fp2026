## Base conceptual de la sesión

### 1. If-Elif-Else
#### Definición
organiza varias alternativas cuando el problema tiene más de dos posibles respuestas.

#### Tipos principales
decisión múltiple con varias ramas ordenadas por prioridad.

#### Ejemplo
clasificar una nota en destacado, esperado o refuerzo.

### 2. Decisiones Anidadas
#### Definición
aparecen cuando una decisión depende del resultado de otra decisión previa.

#### Tipos principales
una decisión interna dentro de una decisión externa.

#### Ejemplo
primero validar rango y luego clasificar el valor.

### 3. Clasificación
#### Definición
consiste en ubicar un dato dentro de una categoría según reglas definidas.

#### Tipos principales
por rangos, categorías o cumplimiento de reglas.

#### Ejemplo
ubicar una temperatura en frío, templado o cálido.

### 4. Prioridad De Reglas
#### Definición
obliga a ordenar las condiciones para que cada caso caiga en la opción correcta.

#### Tipos principales
reglas más restrictivas primero y reglas generales después.

#### Ejemplo
evaluar primero monto >= 500 y luego monto >= 200.

### 5. Casos Límite
#### Definición
son valores cercanos al cambio de regla y ayudan a descubrir errores lógicos.

#### Tipos principales
valores mínimos, máximos y de frontera entre categorías.

#### Ejemplo
probar exactamente 18 años o una nota igual a 14.

## Ejercicio 3

## Resolución guiada aplicada

2. Entradas: identifica las tres entradas
3. Procesos:
- decide el orden de comparación
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- considera casos con igualdad entre valores
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
a = float(input('Número 1: '))
b = float(input('Número 2: '))
c = float(input('Número 3: '))

if a == b == c:
    print('Los tres números son iguales')
elif a >= b and a >= c:
    print('El mayor es:', a)
    if a == b or a == c:
        print('Existe empate en el valor mayor')
elif b >= a and b >= c:
    print('El mayor es:', b)
    if b == a or b == c:
        print('Existe empate en el valor mayor')
else:
    print('El mayor es:', c)
    if c == a or c == b:
        print('Existe empate en el valor mayor')
````

## Tarea tipo competencia

Solicita el monto de una compra y clasifica el descuento aplicable según rangos definidos por el docente.

## Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
monto = float(input('Monto de compra: '))

if monto >= 500:
    descuento = 0.20
elif monto >= 200:
    descuento = 0.10
else:
    descuento = 0.05

print('Descuento aplicado:', descuento * 100, '%')
print('Monto final:', round(monto * (1 - descuento), 2))
````
