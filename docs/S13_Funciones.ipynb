**1. Funciones**
**Concepto**
Agrupan instrucciones con una tarea clara para reutilizar y ordenar la solución.

**Elementos o clasificación**
Funciones sin retorno, con retorno y con parámetros.

**Criterio de uso**
Se usan para dividir un problema en subproblemas. Una buena función tiene nombre claro, responsabilidad única y entradas definidas.

**Ejemplo de aplicación**
```python
def doble(x):
    return x * 2
```

**2. Parámetros**
**Concepto**
Permiten que una función reciba datos desde el exterior para trabajar con ellos.

**Elementos o clasificación**
Parámetros de entrada simples o múltiples.

**Criterio de uso**
Permiten que la función trabaje con datos variables sin depender de valores escritos directamente dentro de ella.

**Ejemplo de aplicación**
```python
def saludar(nombre):
    print(nombre)
```

**3. Return**
**Concepto**
Es la instrucción que entrega al programa principal el resultado producido por una función.

**Elementos o clasificación**
Retorno de un valor calculado, retorno de texto y retorno lógico.

**Criterio de uso**
Devuelve el resultado calculado por una función para que pueda usarse en otra parte del programa.

**Ejemplo de aplicación**
```python
def sumar(a, b):
    return a + b
```

**4. Modularización**
**Concepto**
Divide el problema en partes pequeñas para mejorar claridad y mantenimiento.

**Elementos o clasificación**
División en funciones pequeñas con responsabilidades claras.

**Criterio de uso**
Ayuda a construir programas más claros: cada módulo resuelve una parte y el programa principal coordina el flujo general.

**Ejemplo de aplicación**
Crear una función para registrar y otra para consultar.

**5. Reutilización**
**Concepto**
Consiste en diseñar una solución para usarla más de una vez con distintos datos, evitando repetir código.

**Elementos o clasificación**
Uso repetido de una función con distintos argumentos.

**Criterio de uso**
Una solución reutilizable evita repetir instrucciones y permite aplicar la misma función con datos diferentes.

**Ejemplo de aplicación**
Usar calcular_promedio() para varios estudiantes sin repetir la fórmula.

### Ejercicio 2: Promedio y resultado final

Crearemos una función para calcular el promedio y otra para decidir el resultado final del estudiante.

**Resolución guiada**

1. Comprender: explica con una oración completa cuál es el resultado que debe obtener el programa.
2. Entradas: enumera cada dato de entrada e indica su tipo de dato esperado.
3. Procesos:
- describe la operación, condición, recorrido o función principal que usarás para resolver el problema
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: especifica la salida esperada y el formato con que se debe mostrar.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que el resultado esperado se cumpla
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué sucede con una entrada vacía, inválida o fuera de rango
6. Código:

````python
def calcular_promedio(n1, n2, n3):
    return (n1 + n2 + n3) / 3

def obtener_resultado(promedio):
    if promedio >= 11:
        return 'Aprobado'
    return 'Desaprobado'

promedio = calcular_promedio(15, 12, 14)
print('Promedio:', promedio)
print('Resultado:', obtener_resultado(promedio))
````

### Ejercicio 4: Categoría de desempeño

Escribe una función que reciba un promedio y retorne una categoría de desempeño.

**Resolución guiada**

1. Comprender: una regla de clasificación puede encapsularse en una función reutilizable.
2. Entradas: un promedio como parámetro.
3. Procesos:
- usa condicionales dentro de la función
- retorna un texto según el rango del promedio
- llama la función desde el programa principal
4. Salidas: categoría devuelta por la función.
5. Pruebas: prueba con `19`, `15`, `11` y `8`.
6. Código:

````python
def clasificar_promedio(promedio):
    if promedio >= 18:
        return 'Destacado'
    elif promedio >= 14:
        return 'Bueno'
    elif promedio >= 11:
        return 'Aprobado'
    else:
        return 'En proceso'

promedio = float(input('Promedio: '))
print('Categoría:', clasificar_promedio(promedio))
````

### Ejercicio 5: Mayor valor con función

Escribe una función que reciba una lista de números y retorne el mayor valor. Luego úsala en un programa principal.

**Resolución guiada**

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:

````python
def mayor_lista(numeros):
    mayor = numeros[0]
    for numero in numeros:
        if numero > mayor:
            mayor = numero
    return mayor

datos = [7, 12, 5, 18, 9]
print(mayor_lista(datos))
````
