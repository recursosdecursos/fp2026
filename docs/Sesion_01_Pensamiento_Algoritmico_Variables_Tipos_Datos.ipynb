## Base conceptual de la sesión

### 1. Algoritmos
#### Definición
un algoritmo es una secuencia finita y ordenada de pasos que permite resolver un problema o realizar una tarea. Su valor está en que describe con claridad qué hacer, en qué orden hacerlo y cuándo termina el proceso.

#### Tipos principales
pueden ser secuenciales cuando siguen pasos lineales, condicionales cuando toman decisiones y repetitivos cuando ejecutan acciones varias veces hasta cumplir una condición o un número de repeticiones.

#### Ejemplo
un algoritmo para calcular el área de un rectángulo puede seguir estos pasos: leer base, leer altura, multiplicar ambos valores y mostrar el resultado.

### 2. Entrada, Proceso Y Salida
#### Definición
permite distinguir qué datos ingresan, qué transformación ocurre y qué resultado se obtiene.

#### Tipos principales
entrada de datos, transformación de datos y presentación de resultados.

#### Ejemplo
leer edad, evaluar si es mayor de edad y mostrar el mensaje final.

### 3. Variables
#### Definición
una variable es un nombre asociado a un espacio de memoria donde se guarda un dato que puede usarse y cambiar durante la ejecución del programa. Sirve para representar información del problema de manera comprensible y manipulable.

#### Tipos principales
según el dato que guardan, suelen trabajar con valores numéricos, textos y valores lógicos. También pueden distinguirse por su función dentro del algoritmo, por ejemplo variables de entrada, de proceso o de salida.

#### Ejemplo
nombre = "Ana", edad = 18 y promedio = 16.5 muestran cómo una variable puede representar distintos datos del mismo problema.

### 4. Tipos De Datos
#### Definición
los tipos de datos indican la clase de valor que una variable puede almacenar y la forma en que ese valor será tratado por el programa. Gracias a ellos se distingue si un dato representa texto, números o valores lógicos.

#### Tipos principales
en programación inicial se trabaja sobre todo con enteros, decimales, cadenas de texto y booleanos. Cada tipo define operaciones válidas y una forma distinta de representar la información.

#### Ejemplo
en Python, "Lima" es una cadena, 25 es un entero, 16.5 es un decimal y True es un booleano.

### 5. Salida Formateada
#### Definición
ayuda a presentar la información de forma legible para que el resultado tenga sentido para el usuario.

#### Tipos principales
salida con texto fijo, salida con etiquetas y salida combinada con variables.

#### Ejemplo
print("Nombre:", nombre).

## Ejercicio 2

Construiremos un programa que solicite nombre, edad y carrera de un estudiante y luego muestre una ficha resumida.

## Resolución guiada aplicada

1. Comprender: se debe leer el nombre, la edad y la carrera de un estudiante para mostrar una ficha resumida con esos datos e indicar si es mayor de edad.
2. Entradas: ingresan tres datos por teclado: `nombre` como texto, `edad` como número entero y `carrera` como texto.
3. Procesos:
- leer nombre con input()
- leer edad con input() y convertirla con int()
- leer carrera con input()
- mostrar la ficha ordenada con etiquetas claras
- evaluar la condición edad >= 18 para indicar si es mayor de edad
4. Salidas: debe mostrarse una ficha con las etiquetas `Nombre`, `Edad`, `Carrera` y una indicación sobre si es mayor de edad.
5. Pruebas:
- caso normal: nombre = Ana, edad = 20, carrera = Medicina; debe mostrar la ficha completa y ¿Es mayor de edad?: True
- caso límite: nombre = Luis, edad = 18, carrera = Derecho; debe seguir mostrando True
- caso excepcional: si la edad se ingresa vacía, como texto no numérico o con un valor negativo, el programa actual fallará o producirá un dato inválido, así que ese caso sirve para discutir validación posterior
6. Código:


````python
nombre = input('Nombre: ')
edad = int(input('Edad: '))
carrera = input('Carrera: ')

print('--- Ficha del estudiante ---')
print('Nombre:', nombre)
print('Edad:', edad)
print('Carrera:', carrera)
print('¿Es mayor de edad?:', edad >= 18)
````

## Ejercicio 3

Solicita nombre, ciclo y promedio de un estudiante. Luego muestra un resumen e indica si su promedio está por encima de 14.

## Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Solicita nombre, ciclo y promedio de un estudiante. Luego muestra un resumen e indica si su promedio está por encima de 14
2. Entradas: identifica las tres entradas y el tipo de dato adecuado
3. Procesos:
- define cómo calcularás o evaluarás el promedio ingresado
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: describe qué mensajes debe mostrar el resumen final
5. Pruebas:
- caso normal: usa datos habituales para comprobar que el resultado principal sea correcto
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
nombre = input('Nombre: ')
ciclo = input('Ciclo: ')
promedio = float(input('Promedio: '))

print('--- Resumen del estudiante ---')
print('Nombre:', nombre)
print('Ciclo:', ciclo)
print('Promedio:', promedio)
print('¿Supera 14?:', promedio > 14)
````

## Tarea tipo competencia

Solicita el nombre de un producto, su precio y la cantidad comprada. Luego muestra un resumen con el costo total.

## Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
producto = input('Producto: ')
precio = float(input('Precio: '))
cantidad = int(input('Cantidad: '))

total = precio * cantidad
print('--- Resumen de compra ---')
print('Producto:', producto)
print('Cantidad:', cantidad)
print('Total:', round(total, 2))
````
