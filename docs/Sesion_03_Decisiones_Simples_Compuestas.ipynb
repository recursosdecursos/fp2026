## Ejemplo guiado

Determinaremos si una persona accede a un descuento según su edad y si cuenta con carné estudiantil.


```python
edad = int(input('Edad: '))
tiene_carne = input('¿Tiene carné estudiantil? (si/no): ').strip().lower()

if edad < 18 or tiene_carne == 'si':
    print('Aplica a descuento.')
else:
    print('No aplica a descuento.')
```
