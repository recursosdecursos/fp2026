# Ejercicio 1: Ingresar notas hasta escribir -1

## Resolución guiada aplicada

1. Comprender: no sabemos cuántas notas ingresará el usuario, por eso necesitamos una señal para terminar.
2. Entradas: notas ingresadas una por una hasta escribir `-1`.
3. Proceso: usar `while`, guardar con `append()` y luego calcular con `for`.
4. Salida y pruebas: cantidad de notas y promedio. Prueba con: varias notas; luego probar escribir `-1` al inicio.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
notas = []
nota = float(input("Ingrese nota (-1 para terminar): "))

while nota != -1:
    notas.append(nota)
    nota = float(input("Ingrese nota (-1 para terminar): "))

suma = 0
for valor in notas:
    suma = suma + valor

if len(notas) > 0:
    print("Cantidad de notas:", len(notas))
    print("Promedio:", suma / len(notas))
else:
    print("No se ingresaron notas")
```

# Ejercicio 2: Registrar compras hasta escribir 0

## Resolución guiada aplicada

1. Comprender: el usuario puede registrar varias compras y usará 0 como señal para terminar.
2. Entradas: montos de compra hasta escribir `0`.
3. Proceso: guardar montos en una lista y sumar el total.
4. Salida y pruebas: cantidad de compras y total. Prueba con: `10, 20, 0`; luego `0` al inicio.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
compras = []
monto = float(input("Monto de compra (0 para terminar): "))

while monto != 0:
    compras.append(monto)
    monto = float(input("Monto de compra (0 para terminar): "))

total = 0
for compra in compras:
    total = total + compra

print("Cantidad de compras:", len(compras))
print("Total:", total)
```

# Ejercicio 3: Menú simple

## Resolución guiada aplicada

1. Comprender: el programa debe repetir un menú hasta que el usuario elija salir.
2. Entradas: `opcion` del menú.
3. Proceso: repetir con `while` hasta elegir salir y responder con condicionales.
4. Salida y pruebas: mensaje por opción o error. Prueba con: `1`, `2`, opción inválida y salir.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
opcion = ""

while opcion != "3":
    print("1. Saludar")
    print("2. Mostrar mensaje")
    print("3. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        print("Hola")
    elif opcion == "2":
        print("Sigue practicando Python")
    elif opcion == "3":
        print("Fin del programa")
    else:
        print("Opción inválida")
```
