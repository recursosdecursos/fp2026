# Ejercicio 1: Sumar números del 1 a N

## Resolución guiada aplicada

1. Comprender: necesitamos sumar una secuencia de números desde 1 hasta el número que indique el usuario.
2. Entradas: `n`.
3. Proceso: recorrer del 1 a `n` con `for` y acumular la suma.
4. Salida y pruebas: suma total. Prueba con: `n = 5` debe dar 15; `n = 1` debe dar 1.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
n = int(input("Ingrese N: "))
suma = 0

for numero in range(1, n + 1):
    suma = suma + numero

print("La suma es:", suma)
```

# Ejercicio 2: Contar pares en una lista

## Resolución guiada aplicada

1. Comprender: tenemos varios números y queremos saber cuántos de ellos son pares.
2. Entradas: lista de números.
3. Proceso: recorrer la lista y aumentar un contador cuando `numero % 2 == 0`.
4. Salida y pruebas: cantidad de pares. Prueba con: lista con pares e impares; lista sin pares.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
numeros = [4, 7, 10, 15, 20, 21]
contador_pares = 0

for numero in numeros:
    if numero % 2 == 0:
        contador_pares = contador_pares + 1

print("Cantidad de pares:", contador_pares)
```

# Ejercicio 3: Promedio de una lista de notas

## Resolución guiada aplicada

1. Comprender: cuando hay muchas notas, conviene guardarlas en una lista y calcular el promedio recorriéndola.
2. Entradas: lista de notas.
3. Proceso: recorrer la lista, acumular suma y dividir entre `len(notas)`.
4. Salida y pruebas: promedio. Prueba con: `[15, 15, 15]`; `[20, 10]`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
notas = [15, 18, 13, 16]
suma = 0

for nota in notas:
    suma = suma + nota

promedio = suma / len(notas)
print("Promedio:", promedio)
```
