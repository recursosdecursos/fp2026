# Ejercicio 1: Registro de estudiante con diccionario

## Resolución guiada aplicada

1. Comprender: los datos de un estudiante pertenecen a una misma entidad; un diccionario ayuda a organizarlos por nombre.
2. Entradas: datos con nombre: por ejemplo `nombre`, `grado` y `nota`.
3. Proceso: guardar cada dato con una clave dentro de un diccionario.
4. Salida y pruebas: valores recuperados por clave. Prueba con: un estudiante con nota decimal y otro con nota entera.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
estudiante = {
    "nombre": input("Nombre: "),
    "grado": input("Grado: "),
    "nota": float(input("Nota: "))
}

print("Nombre:", estudiante["nombre"])
print("Grado:", estudiante["grado"])
print("Nota:", estudiante["nota"])
```

# Ejercicio 2: Tabla de notas con lista de listas

## Resolución guiada aplicada

1. Comprender: una tabla permite guardar varias filas, donde cada fila representa un estudiante y su nota.
2. Entradas: tabla con filas de datos.
3. Proceso: recorrer cada fila y acceder por posición.
4. Salida y pruebas: nombre y nota de cada estudiante. Prueba con: agrega una fila nueva y vuelve a ejecutar.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
tabla = [
    ["Ana", 17],
    ["Luis", 14],
    ["María", 19]
]

for fila in tabla:
    print(fila[0], "tiene", fila[1])
```

# Ejercicio 3: Lista de diccionarios

## Resolución guiada aplicada

1. Comprender: cuando tenemos varios objetos con datos nombrados, podemos guardar varios diccionarios dentro de una lista.
2. Entradas: datos con nombre: por ejemplo `nombre`, `grado` y `nota`.
3. Proceso: guardar cada dato con una clave dentro de un diccionario.
4. Salida y pruebas: valores recuperados por clave. Prueba con: un estudiante con nota decimal y otro con nota entera.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
productos = [
    {"nombre": "Cuaderno", "precio": 5.5},
    {"nombre": "Lapicero", "precio": 1.2},
    {"nombre": "Mochila", "precio": 80.0}
]

for producto in productos:
    print(producto["nombre"], "cuesta", producto["precio"])
```
