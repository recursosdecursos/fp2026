## Objetivo

## Contenidos

## Base conceptual de la sesión

### 1. For
#### Definición
repite una acción un número conocido de veces.

#### Tipos principales
recorrido por contador, por rango y por colección.

#### Ejemplo
for i in range(5): leer una nota.

### 2. Range
#### Definición
genera una secuencia de valores útil para controlar repeticiones con for.

#### Tipos principales
inicio-fin, inicio-fin-paso y rango simple desde cero.

#### Ejemplo
range(1, 6) genera cinco valores consecutivos.

### 3. Contadores
#### Definición
registran cuántas veces ocurre una condición o evento.

#### Tipos principales
conteo simple, conteo por condición y conteo acumulado por iteración.

#### Ejemplo
aprobados += 1 cuando la nota cumple la condición.

### 4. Acumuladores
#### Definición
guardan una suma u otro resultado que crece durante el ciclo.

#### Tipos principales
sumatorias, productos y agregaciones progresivas.

#### Ejemplo
suma += numero en cada repetición.

### 5. Sumatorias
#### Definición
expresan la acumulación progresiva de valores dentro de un proceso repetitivo.

#### Tipos principales
sumatoria total, parcial y condicionada.

#### Ejemplo
sumar seis números ingresados por el usuario.

## Ejercicio 1

Contaremos cuántos estudiantes aprueban dentro de un grupo de cinco registros ingresados por teclado.

## Resolución guiada aplicada

1. Comprender: explica con una oración completa cuál es el resultado que debe obtener el programa.
2. Entradas: enumera cada dato de entrada e indica su tipo de dato esperado.
3. Procesos:

- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: especifica la salida esperada y el formato con que se debe mostrar.

- caso excepcional: revisa qué sucede con una entrada vacía, inválida o fuera de rango
6. Código:


````python
aprobados = 0
for i in range(5):
    nota = float(input(f'Nota {i + 1}: '))
    if nota >= 11:
        aprobados += 1

print('Total de aprobados:', aprobados)
````

## Ejercicio 2

Solicita 5 edades y muestra cuántas corresponden a personas mayores de edad.

## Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Solicita 5 edades y muestra cuántas corresponden a personas mayores de edad
2. Entradas: determina cuántas repeticiones tendrá el ciclo
3. Procesos:
- define cuándo aumenta el contador
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- prueba con edades mezcladas

- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
mayores = 0
for i in range(5):
    edad = int(input(f'Edad {i + 1}: '))
    if edad >= 18:
        mayores += 1

print('Mayores de edad:', mayores)
````

## Ejercicio 3

Solicita 4 precios y calcula el total acumulado y el promedio de compra.

## Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Solicita 4 precios y calcula el total acumulado y el promedio de compra
2. Entradas: define el acumulador principal
3. Procesos:
- repite la lectura exactamente 4 veces. calcula el promedio al final
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que el resultado principal sea correcto
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
total = 0
for i in range(4):
    precio = float(input(f'Precio {i + 1}: '))
    total += precio

promedio = total / 4
print('Total:', round(total, 2))
print('Promedio:', round(promedio, 2))
````

## Tarea tipo competencia

Solicita 6 números y muestra la suma total y cuántos son pares.

## Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
suma = 0
pares = 0

for i in range(6):
    numero = int(input(f'Número {i + 1}: '))
    suma += numero
    if numero % 2 == 0:
        pares += 1

print('Suma:', suma)
print('Pares:', pares)
````
