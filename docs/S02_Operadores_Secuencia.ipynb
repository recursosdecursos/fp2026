## Base conceptual de la sesión

### 1. Operadores Aritméticos
#### Concepto
Permiten realizar cálculos como suma, resta, multiplicación, división y residuos.

#### Elementos o clasificación
Suma, resta, multiplicación, división, división entera, potencia y módulo.

#### Criterio de uso
Se usan cuando el proceso requiere transformar datos numéricos. Es importante cuidar la precedencia y usar paréntesis cuando la expresión pueda ser ambigua.

#### Ejemplo de aplicación
Promedio = (n1 + n2 + n3) / 3.

### 2. Operadores Relacionales
#### Concepto
Comparan valores y producen resultados lógicos como verdadero o falso.

#### Elementos o clasificación
Igualdad, diferencia, mayor que, menor que, mayor o igual y menor o igual.

#### Criterio de uso
Se emplean para formular preguntas sobre los datos. El resultado de una comparación alimenta decisiones, validaciones y filtros.

#### Ejemplo de aplicación
Promedio >= 11.

### 3. Operadores Lógicos
#### Concepto
Combinan condiciones para construir decisiones más completas.

#### Elementos o clasificación
Conjunción, disyunción y negación.

#### Criterio de uso
Permiten unir condiciones simples. Deben usarse con cuidado porque cambian el significado de la decisión: and exige que todo se cumpla, or acepta alternativas y not invierte la condición.

#### Ejemplo de aplicación
Edad < 18 or tiene_carne == "si".

### 4. Expresiones
#### Concepto
Integran datos, operadores y variables para producir un resultado nuevo.

#### Elementos o clasificación
Aritméticas, relacionales y lógicas.

#### Criterio de uso
Una expresión debe ser legible y verificable. Si combina varios cálculos, puede dividirse en variables intermedias para mejorar claridad.

#### Ejemplo de aplicación
Total = precio * cantidad.

### 5. Secuencia De Instrucciones
#### Concepto
Organiza el programa en un orden de ejecución donde cada paso depende del anterior.

#### Elementos o clasificación
Lectura, cálculo y salida en orden lineal.

#### Criterio de uso
En una estructura secuencial, el orden importa: primero se leen o definen datos, luego se calculan resultados y finalmente se muestran las salidas.

#### Ejemplo de aplicación
Leer datos, calcular y luego mostrar.

### Ejercicio 2

Solicita cuatro notas, calcula el promedio y muestra también la nota mayor ingresada.

#### Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Solicita cuatro notas, calcula el promedio y muestra también la nota mayor ingresada
2. Entradas: anota las cuatro entradas numéricas
3. Procesos:
- define la fórmula del promedio. decide cómo compararás las notas para hallar la mayor
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que el resultado principal sea correcto
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
n1 = float(input('Nota 1: '))
n2 = float(input('Nota 2: '))
n3 = float(input('Nota 3: '))
n4 = float(input('Nota 4: '))

promedio = (n1 + n2 + n3 + n4) / 4
mayor = max(n1, n2, n3, n4)

print('Promedio:', round(promedio, 2))
print('Nota mayor:', mayor)
````

### Ejercicio 3

Pide el precio de un producto y el porcentaje de descuento. Luego calcula el precio final y muestra cuánto se ahorró el cliente.

#### Resolución guiada aplicada

1. Comprender: el programa debe resolver este problema: Pide el precio de un producto y el porcentaje de descuento. Luego calcula el precio final y muestra cuánto se ahorró el cliente
2. Entradas: define las entradas numéricas
3. Procesos:
- escribe la operación para hallar descuento y precio final
- ordena las operaciones o decisiones antes de escribir el código
- prepara el resultado final que mostrarás al usuario
4. Salidas: muestra el resultado solicitado por el enunciado con mensajes claros y ordenados.
5. Pruebas:
- comprueba con un caso de descuento 0 y otro mayor
- caso límite: prueba el valor exacto donde cambia la regla, condición o cálculo principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
precio = float(input('Precio: '))
descuento = float(input('Descuento %: '))

ahorro = precio * descuento / 100
precio_final = precio - ahorro

print('Ahorro:', round(ahorro, 2))
print('Precio final:', round(precio_final, 2))
````

### Tarea tipo competencia

Pide el sueldo base y el porcentaje de bonificación. Luego calcula el sueldo final y muestra si supera un monto de referencia.

#### Resolución guiada aplicada

1. Comprender: redacta con tus palabras qué debe resolver la tarea.
2. Entradas: define qué datos ingresan y el tipo que tendrán.
3. Procesos:
- describe la estrategia, fórmula, condición, estructura o algoritmo que piensas aplicar
- ordena las acciones principales antes de escribir el código
- prepara cómo obtendrás el resultado final antes de mostrarlo
4. Salidas: indica con claridad qué resultado debe mostrar o registrar la solución.
5. Pruebas:
- caso normal: usa datos habituales para comprobar que la solución funcione
- caso límite: prueba el valor exacto donde cambia la regla principal
- caso excepcional: revisa qué ocurre con una entrada vacía, inválida o fuera de rango
6. Código:


````python
sueldo_base = float(input('Sueldo base: '))
porcentaje_bono = float(input('Porcentaje de bonificación: '))

bono = sueldo_base * porcentaje_bono / 100
sueldo_final = sueldo_base + bono
print('Sueldo final:', round(sueldo_final, 2))
print('¿Supera 2000?:', sueldo_final > 2000)
````
