### Ejercicio 1: Descuento en una compra

Construir un programa que lea el monto de una compra. Si el monto es mayor o igual que `100`, aplica un descuento del `10%`; de lo contrario, no aplica descuento.

**ResoluciÃ³n guiada**

1. Comprender: el total a pagar cambia segÃºn el monto de compra.
2. Entradas: `monto`, como nÃºmero decimal.
3. Procesos:
- calcular el descuento solo cuando corresponde
- asignar el porcentaje aplicado
- calcular el total final despuÃ©s del descuento
4. Salidas: descuento aplicado, monto descontado y total a pagar.
5. Pruebas: usa `80`, `100` y `150`.
6. CÃ³digo:

````python
monto = float(input('Monto de compra: '))

if monto >= 100:
    porcentaje_descuento = 10
    descuento = monto * 0.10
else:
    porcentaje_descuento = 0
    descuento = 0

total = monto - descuento

print('Descuento aplicado:', porcentaje_descuento, '%')
print('Monto descontado:', descuento)
print('Total a pagar:', total)
````

Construir un programa que lea un nÃºmero entero y muestre su valor absoluto sin usar la funciÃ³n `abs()`.

### Ejercicio 5: DuraciÃ³n de un juego

Reto tipo competencia. Un juego inicia en una hora entera y termina en otra hora entera del mismo dÃ­a o del dÃ­a siguiente. Construir un programa que calcule cuÃ¡ntas horas durÃ³ el juego.

**Problema**

Se leen dos enteros: `inicio` y `fin`, ambos entre `0` y `23`.

**Reglas**

- Si la hora final es mayor que la hora inicial, el juego termina el mismo dÃ­a.
- Si la hora final es menor o igual que la hora inicial, el juego termina al dÃ­a siguiente.
- La duraciÃ³n mÃ­nima es `1` hora y la duraciÃ³n mÃ¡xima es `24` horas.

**Entrada**

- Hora de inicio.
- Hora de fin.

**Salida esperada**

Mostrar la duraciÃ³n exacta en horas.

**ResoluciÃ³n guiada**

1. Comprender: el cambio de dÃ­a obliga a decidir cÃ³mo calcular la diferencia.
2. Entradas: `inicio` y `fin`, como enteros.
3. Procesos:
- si `fin > inicio`, calcular `fin - inicio`
- en caso contrario, calcular `24 - inicio + fin`
- guardar la duraciÃ³n antes de mostrarla
4. Salidas: duraciÃ³n del juego en horas.
5. Pruebas: usa `10` y `12`; `22` y `2`; `8` y `8`.
6. CÃ³digo:

````python
inicio = int(input('Hora de inicio: '))
fin = int(input('Hora de fin: '))

if fin > inicio:
    duracion = fin - inicio
else:
    duracion = 24 - inicio + fin

print('DuraciÃ³n del juego:', duracion, 'hora(s)')
````
