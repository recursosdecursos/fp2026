**1. Algoritmos condicionales**

**Concepto**
Un algoritmo condicional es aquel que evalúa una condición para elegir entre uno o más caminos posibles. A diferencia de un algoritmo secuencial, no todas las instrucciones se ejecutan siempre; algunas dependen del resultado de una pregunta lógica.

**Diagrama de flujo**

```mermaid
flowchart TD
inicio([Inicio])
entrada[/Leer datos/]
proceso[Procesar datos]
decision{¿Cumple condición?}
accion_si[Ejecutar acciones del caso verdadero]
accion_no[Ejecutar acciones del caso falso]
fin([Fin])
inicio --> entrada --> proceso --> decision
decision -->|Sí| accion_si --> fin
decision -->|No| accion_no --> fin
```

**Representación en texto**

```text
                 +----------------+
                 |     Inicio     |
                 +----------------+
                         |
                         v
                 +----------------+
                 |   Leer datos   |
                 +----------------+
                         |
                         v
                 +----------------+
                 | Procesar datos |
                 +----------------+
                         |
                         v
              +----------------------+
              | ¿Cumple condición?   |
              +----------------------+
                    /          \
                  Sí            No
                  /              \
                 v                v
+--------------------------+   +--------------------------+
| Acciones caso verdadero |   | Acciones caso falso     |
+--------------------------+   +--------------------------+
                 \              /
                  \            /
                   v          v
                 +----------------+
                 |      Fin       |
                 +----------------+
```

**2. If**
**Concepto**
Permite ejecutar una acción solo cuando una condición se cumple.

**Elementos o clasificación**
Decisión simple con una sola condición y una sola acción principal.

**Criterio de uso**
Se usa cuando una acción depende de una condición. Si la condición es falsa y no existe otra rama, el programa continúa con la siguiente instrucción.

**Ejemplo de aplicación**
```python
if nota >= 11:
    print("Aprueba")
```

**3. If-Else**
**Concepto**
Define dos caminos posibles según el resultado de una condición.

**Elementos o clasificación**
Decisión binaria con camino verdadero y camino falso.

**Criterio de uso**
Se usa cuando el problema exige escoger entre dos caminos excluyentes. Cada camino debe producir un resultado coherente para el usuario.

**Ejemplo de aplicación**
```python
if edad >= 18:
    print("Mayor")
else:
    print("Menor")
```

**4. Condiciones Compuestas**
**Concepto**
Usan operadores lógicos para evaluar más de una regla al mismo tiempo.

**Elementos o clasificación**
Condiciones unidas por and, or y not.

**Criterio de uso**
Conviene construirlas a partir de condiciones simples ya entendidas. Si una condición se vuelve extensa, debe separarse en variables lógicas con nombres claros.

**Ejemplo de aplicación**
```python
if edad >= 18 and promedio >= 14:
    print("Beca")
```

**5. Validación Básica**
**Concepto**
Sirve para comprobar si los datos ingresados tienen sentido antes de continuar.

**Elementos o clasificación**
Validación de rango, formato y consistencia.

**Criterio de uso**
La validación protege al algoritmo de datos imposibles o incoherentes antes de realizar cálculos, clasificaciones o salidas finales.

**Ejemplo de aplicación**
```python
if nota < 0 or nota > 20:
    print("Fuera de rango")
```

**6. Casos De Prueba**
**Concepto**
Permiten verificar si la solución funciona en situaciones normales, límite y excepcionales.

**Elementos o clasificación**
Casos normales, límite y excepcionales.

**Criterio de uso**
Una solución debe probarse con datos normales, valores de frontera y entradas inválidas o poco frecuentes.

**Ejemplo de aplicación**
Probar con 10.9, 11 y 20 para observar cambios de resultado.

### Ejercicio 1: Aprobación por promedio y asistencia

Pide el promedio final y la asistencia de un estudiante. Determina si aprueba solo si el promedio es al menos `11` y la asistencia supera `70%`.

**Resolución guiada**

1. Comprender: deben cumplirse dos condiciones al mismo tiempo.
2. Entradas: `promedio` y `asistencia`, como números decimales.
3. Procesos:
- comparar el promedio con `11`
- comparar la asistencia con `70`
- usar `and` para combinar las reglas
- asignar estado, observación y puntos según el resultado
4. Salidas: estado, observación y puntos reconocidos.
5. Pruebas: usa un caso que apruebe, uno que falle por promedio y otro que falle por asistencia.
6. Código:

````python
promedio = float(input('Promedio: '))
asistencia = float(input('Asistencia %: '))

if promedio >= 11 and asistencia > 70:
    estado = 'Aprobado'
    observacion = 'Cumple promedio y asistencia'
    puntos = promedio
else:
    estado = 'Desaprobado'
    observacion = 'Debe revisar promedio o asistencia'
    puntos = 0

print('Estado:', estado)
print('Observación:', observacion)
print('Puntos reconocidos:', puntos)
````

### Ejercicio 2: Validación de nota

Solicita una nota y valida si está dentro del rango permitido de 0 a 20.

**Resolución guiada**

1. Comprender: una nota solo es válida si cumple el límite inferior y superior.
2. Entradas: una nota numérica.
3. Procesos:
- lee la nota con `float()`
- usa `and` para verificar `nota >= 0` y `nota <= 20`
- si la nota es válida, conservarla como nota final
- si la nota es inválida, registrar `0` como nota final
4. Salidas: mensaje de validación y nota final registrada.
5. Pruebas: prueba con `0`, `20`, `21` y `-1`.
6. Código:

````python
nota = float(input('Nota: '))

if nota >= 0 and nota <= 20:
    nota_final = nota
    mensaje = 'Nota válida'
else:
    nota_final = 0
    mensaje = 'Nota inválida: se registrará 0'

print(mensaje)
print('Nota final registrada:', nota_final)
````

### Ejercicio 3: Valor absoluto

Construir un programa que lea un número entero y muestre su valor absoluto sin usar la función `abs()`.

**Resolución guiada**

1. Comprender: el valor absoluto representa la distancia de un número respecto de cero.
2. Entradas: `numero`, como entero.
3. Procesos:
- si el número es negativo, multiplicarlo por `-1`
- si el número no es negativo, conservarlo igual
- mostrar el resultado final
4. Salidas: valor absoluto del número.
5. Pruebas: usa `-8`, `0` y `12`.
6. Código:

````python
numero = int(input('Número: '))

if numero < 0:
    valor_absoluto = numero * -1
else:
    valor_absoluto = numero

print('Valor absoluto:', valor_absoluto)
````

### Ejercicio 4: Evaluación de envío

En una competencia de programación, los enunciados suelen pedir una salida exacta. En este reto, debes determinar si un pedido puede enviarse gratis o si debe pagar envío.

**Problema**

Una tienda evalúa un pedido usando tres datos: monto de compra, distancia de entrega y si el cliente es frecuente.

**Reglas**

- El pedido tiene envío gratis si el monto es mayor o igual que `100` y la distancia es menor o igual que `5`.
- También tiene envío gratis si el cliente es frecuente y el monto es mayor o igual que `60`.
- En cualquier otro caso, debe pagar envío.

**Entrada**

- Monto de compra como número decimal.
- Distancia en kilómetros como número decimal.
- Cliente frecuente como texto: `si` o `no`.

**Salida esperada**

Mostrar el estado del envío y el total final. El estado debe ser uno de estos valores:

```text
ENVIO GRATIS
```

```text
PAGA ENVIO
```

**Resolución guiada**

1. Comprender: existen dos formas de obtener envío gratis; si ninguna se cumple, se paga envío.
2. Entradas: `monto`, `distancia` y `frecuente`.
3. Procesos:
- construir la primera condición con `and`
- construir la segunda condición con `and`
- unir ambas posibilidades con `or`
- usar `if-else` para asignar costo de envío, total y estado
4. Salidas: estado, costo de envío y total final.
5. Pruebas: usa un caso que cumpla la primera regla, otro que cumpla la segunda y otro que no cumpla ninguna.
6. Código:

````python
monto = float(input('Monto: '))
distancia = float(input('Distancia: '))
frecuente = input('Cliente frecuente (si/no): ').strip().lower()

regla_monto_distancia = monto >= 100 and distancia <= 5
regla_frecuente = frecuente == 'si' and monto >= 60

envio_gratis = regla_monto_distancia or regla_frecuente

if envio_gratis:
    costo_envio = 0
    total = monto
    estado = 'ENVIO GRATIS'
else:
    costo_envio = 12
    total = monto + costo_envio
    estado = 'PAGA ENVIO'

print('ESTADO:', estado)
print('COSTO_ENVIO:', costo_envio)
print('TOTAL:', round(total, 2))
````
