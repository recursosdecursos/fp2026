**1. Operadores aritméticos**

Permiten realizar cálculos con datos numéricos.

Los operadores más usados son:

- `+` suma
- `-` resta
- `*` multiplicación
- `/` división decimal
- `//` división entera
- `%` residuo
- `**` potencia

Ejemplo:

```python
total = precio * cantidad
sobrante = caramelos % estudiantes
```

**2. Expresiones**

Una expresión combina valores, variables y operadores para obtener un nuevo resultado.

```python
promedio = (nota1 + nota2 + nota3) / 3
```

En este caso, el programa primero suma las notas y luego divide entre tres. Los paréntesis ayudan a indicar el orden correcto del cálculo.

**3. Operadores de asignación**

El operador `=` guarda un valor en una variable.

```python
subtotal = precio * cantidad
```

También existen formas abreviadas como `+=`, `-=`, `*=`, pero en esta sesión se usará principalmente la asignación directa para mantener los algoritmos claros.

**4. Precedencia de operadores**

Python evalúa algunas operaciones antes que otras. Por ejemplo, la multiplicación se resuelve antes que la suma.

```python
resultado = 2 + 3 * 4
```

El resultado es `14`, no `20`. Para cambiar el orden se usan paréntesis.

```python
resultado = (2 + 3) * 4
```

**5. Algoritmos secuenciales**

Un algoritmo secuencial ejecuta instrucciones una después de otra, sin decisiones ni repeticiones.

La forma básica es:

```text
Entrada
|
Proceso
|
Salida
```

En esta sesión los problemas se resolverán leyendo datos, calculando resultados y mostrando una salida ordenada.

### Ejercicio 1: Promedio de notas

Construir un programa que lea tres notas, calcule el promedio y lo muestre con dos decimales.

**Resolución guiada**

1. Comprender: se necesita obtener un solo promedio a partir de tres notas.
2. Entradas: `nota1`, `nota2` y `nota3`, como números decimales.
3. Proceso:
- sumar las tres notas
- dividir la suma entre `3`
- redondear el resultado al mostrarlo
4. Salida: promedio final.
5. Pruebas: usa notas como `14`, `16` y `13` para verificar el cálculo.
6. Código:

````python
nota1 = float(input('Nota 1: '))
nota2 = float(input('Nota 2: '))
nota3 = float(input('Nota 3: '))

promedio = (nota1 + nota2 + nota3) / 3

print('Promedio:', round(promedio, 2))
````

### Ejercicio 2: Área y perímetro de un rectángulo

Construir un programa que lea la base y la altura de un rectángulo. Luego debe calcular el área y el perímetro.

**Resolución guiada**

1. Comprender: se deben obtener dos resultados a partir de la base y la altura.
2. Entradas: `base` y `altura`, como números decimales.
3. Proceso:
- calcular `area = base * altura`
- calcular `perimetro = 2 * (base + altura)`
4. Salidas: área y perímetro.
5. Pruebas: usa `base = 8` y `altura = 5`.
6. Código:

````python
base = float(input('Base: '))
altura = float(input('Altura: '))

area = base * altura
perimetro = 2 * (base + altura)

print('Área:', round(area, 2))
print('Perímetro:', round(perimetro, 2))
````

**Código equivalente en Java**

```java

        System.out.print("Altura: ");
        double altura = Double.parseDouble(sc.nextLine());

        double area = base * altura;
        double perimetro = 2 * (base + altura);

        System.out.printf("Área: %.2f%n", area);
        System.out.printf("Perímetro: %.2f%n", perimetro);
    }
}
```

### Ejercicio 3: Boleta de venta

Construir un programa que lea el precio unitario y la cantidad comprada. Luego debe calcular el subtotal, el IGV y el total a pagar.

**Resolución guiada**

1. Comprender: una boleta requiere calcular importes en orden.
2. Entradas: `precio_unitario` como decimal y `cantidad` como entero.
3. Proceso:
- calcular el subtotal multiplicando precio por cantidad
- calcular el IGV como el 18% del subtotal
- calcular el total sumando subtotal e IGV
4. Salidas: subtotal, IGV y total.
5. Pruebas: usa `precio = 25.50` y `cantidad = 3`.
6. Código:

````python
precio_unitario = float(input('Precio unitario: '))
cantidad = int(input('Cantidad: '))

subtotal = precio_unitario * cantidad
igv = subtotal * 0.18
total = subtotal + igv

print('Subtotal:', round(subtotal, 2))
print('IGV:', round(igv, 2))
print('Total:', round(total, 2))
````

**Código equivalente en Java**

```java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Precio unitario: ");
        double precioUnitario = Double.parseDouble(sc.nextLine());

        System.out.print("Cantidad: ");
        int cantidad = Integer.parseInt(sc.nextLine());

        double subtotal = precioUnitario * cantidad;
        double igv = subtotal * 0.18;
        double total = subtotal + igv;

        System.out.printf("Subtotal: %.2f%n", subtotal);
        System.out.printf("IGV: %.2f%n", igv);
        System.out.printf("Total: %.2f%n", total);
    }
}
```

### Ejercicio 4: Reparto de entradas

Construir un programa que lea la cantidad total de entradas y la cantidad de estudiantes. Luego debe mostrar cuántas entradas recibe cada estudiante y cuántas sobran.

**Resolución guiada**

1. Comprender: se debe repartir una cantidad entera en partes iguales.
2. Entradas: `entradas` y `estudiantes`, como números enteros.
3. Proceso:
- usar `//` para calcular cuántas entradas recibe cada estudiante
- usar `%` para calcular cuántas entradas sobran
4. Salidas: entradas por estudiante y entradas sobrantes.
5. Pruebas: usa `25` entradas y `6` estudiantes.
6. Código:

````python
entradas = int(input('Entradas: '))
estudiantes = int(input('Estudiantes: '))

por_estudiante = entradas // estudiantes
sobrantes = entradas % estudiantes

print('Cada estudiante recibe:', por_estudiante)
print('Sobran:', sobrantes)
````

**Código equivalente en Java**

```java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Entradas: ");
        int entradas = Integer.parseInt(sc.nextLine());

        System.out.print("Estudiantes: ");
        int estudiantes = Integer.parseInt(sc.nextLine());

        int porEstudiante = entradas / estudiantes;
        int sobrantes = entradas % estudiantes;

        System.out.println("Cada estudiante recibe: " + porEstudiante);
        System.out.println("Sobran: " + sobrantes);
    }
}
```

### Ejercicio 5: Costo de viaje

Construir un programa que lea la distancia de un viaje en kilómetros, el rendimiento del vehículo en kilómetros por galón y el precio del galón. Luego debe calcular cuántos galones se necesitan y el costo total del viaje.

**Resolución guiada**

1. Comprender: se debe calcular consumo y costo usando una secuencia de fórmulas.
2. Entradas: `distancia`, `rendimiento` y `precio_galon`, como números decimales.
3. Proceso:
- calcular los galones necesarios dividiendo distancia entre rendimiento
- calcular el costo multiplicando galones por precio del galón
4. Salidas: galones necesarios y costo total.
5. Pruebas: usa `distancia = 180`, `rendimiento = 45` y `precio_galon = 17.5`.
6. Código:

````python
distancia = float(input('Distancia en km: '))
rendimiento = float(input('Km por galón: '))
precio_galon = float(input('Precio por galón: '))

galones = distancia / rendimiento
costo = galones * precio_galon

print('Galones necesarios:', round(galones, 2))
print('Costo total:', round(costo, 2))
````

**Código equivalente en Java**

```java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Distancia en km: ");
        double distancia = Double.parseDouble(sc.nextLine());

        System.out.print("Km por galón: ");
        double rendimiento = Double.parseDouble(sc.nextLine());

        System.out.print("Precio por galón: ");
        double precioGalon = Double.parseDouble(sc.nextLine());

        double galones = distancia / rendimiento;
        double costo = galones * precioGalon;

        System.out.printf("Galones necesarios: %.2f%n", galones);
        System.out.printf("Costo total: %.2f%n", costo);
    }
}
```
