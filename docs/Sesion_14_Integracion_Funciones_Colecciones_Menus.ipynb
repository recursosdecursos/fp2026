## Ejemplo guiado

Construiremos un menú básico para registrar y mostrar nombres en una lista.


```python
def mostrar_menu():
    print('1. Agregar nombre')
    print('2. Mostrar nombres')
    print('3. Salir')

nombres = []
opcion = ''

while opcion != '3':
    mostrar_menu()
    opcion = input('Opción: ')
    if opcion == '1':
        nombres.append(input('Nombre: '))
    elif opcion == '2':
        print(nombres)
```
