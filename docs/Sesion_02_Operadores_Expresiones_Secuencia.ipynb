## Ejemplo guiado

Calcularemos el promedio de tres notas y determinaremos si el estudiante aprueba con una condición simple.


```python
nota1 = float(input('Nota 1: '))
nota2 = float(input('Nota 2: '))
nota3 = float(input('Nota 3: '))

promedio = (nota1 + nota2 + nota3) / 3
print('Promedio:', round(promedio, 2))
print('Aprueba:', promedio >= 11)
```
