# Ejercicio 1: Función para sumar


```python
def sumar(a, b):
    return a + b

numero1 = float(input("Primer número: "))
numero2 = float(input("Segundo número: "))
print("Suma:", sumar(numero1, numero2))
```

# Ejercicio 2: Función para promedio


```python
def calcular_promedio(notas):
    suma = 0
    for nota in notas:
        suma = suma + nota
    return suma / len(notas)

notas = [15, 16, 18]
print("Promedio:", calcular_promedio(notas))
```

# Ejercicio 3: Función para clasificar promedio


```python
def clasificar_promedio(promedio):
    if promedio >= 18:
        return "Logro destacado"
    elif promedio >= 14:
        return "Logro esperado"
    elif promedio >= 11:
        return "En proceso"
    else:
        return "En inicio"

promedio = float(input("Promedio: "))
print(clasificar_promedio(promedio))
```

## Ampliación guiada: recursividad


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
