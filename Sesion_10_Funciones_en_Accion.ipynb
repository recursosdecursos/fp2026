# Ejercicio 1: Función para sumar

## Resolución guiada aplicada

1. Comprender: si una operación se repite, conviene convertirla en una función reutilizable.
2. Entradas: `a` y `b` como parámetros.
3. Proceso: la función retorna `a + b`.
4. Salida y pruebas: resultado devuelto por la función. Prueba con: `5` y `3`; `-2` y `10`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
def sumar(a, b):
    return a + b

número1 = float(input("Primer número: "))
número2 = float(input("Segundo número: "))
print("Suma:", sumar(número1, número2))
```

# Ejercicio 2: Función para promedio

## Resolución guiada aplicada

1. Comprender: calcular promedios puede repetirse muchas veces; una función evita reescribir la misma lógica.
2. Entradas: lista de notas como parámetro.
3. Proceso: sumar notas y dividir entre la cantidad.
4. Salida y pruebas: promedio retornado. Prueba con: `[10, 20]`; `[15, 15, 15]`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
def calcular_promedio(notas):
    suma = 0
    for nota in notas:
        suma = suma + nota
    return suma / len(notas)

notas = [15, 16, 18]
print("Promedio:", calcular_promedio(notas))
```

# Ejercicio 3: Función para clasificar promedio

## Resolución guiada aplicada

1. Comprender: la clasificación de un promedio es una regla que podemos encapsular en una función.
2. Entradas: `promedio` como parámetro.
3. Proceso: usarácondiciónales y retornar una categoría.
4. Salida y pruebas: texto de clasificación. Prueba con: `19`, `15`, `11` y `8`.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
def clasificar_promedio(promedio):
    if promedio >= 18:
        return "Logro destácado"
    elif promedio >= 14:
        return "Logro esperado"
    elif promedio >= 11:
        return "En proceso"
    else:
        return "En inicio"

promedio = float(input("Promedio: "))
print(clasificar_promedio(promedio))
```

## Ampliación guiada: recursividad


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```
