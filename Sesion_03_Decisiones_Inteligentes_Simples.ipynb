# Ejercicio 1: Aprobó o desaprobó

## 1. Problema
Saber si un estudiante aprobó según su nota.

## 2. Entrada
- nota

## 3. Proceso
- Si nota >= 11, mostrar que aprobó.
- En caso contrario, mostrar que desaprobó.

## 4. Salida
- Mostrar el resultado.

## 5. Código en Python

````python
nota = float(input("Ingrese su nota: "))

if nota >= 11:
    print("Aprobaste")
else:
    print("Desaprobaste")
````

# Ejercicio 2: Comparar dos números

## 1. Problema
Comparar dos números y mostrar cuál es mayor.

## 2. Entrada
- numero1
- numero2

## 3. Proceso
- Si numero1 > numero2, numero1 es mayor.
- Si numero2 > numero1, numero2 es mayor.
- Si no ocurre lo anterior, son iguales.

## 4. Salida
- Mostrar el mayor o indicar que son iguales.

## 5. Código en Python

````python
numero1 = float(input("Ingrese el primer número: "))
numero2 = float(input("Ingrese el segundo número: "))

if numero1 > numero2:
    print("El mayor es:", numero1)
elif numero2 > numero1:
    print("El mayor es:", numero2)
else:
    print("Los números son iguales")
````

# Ejercicio 3: Condición compuesta

## 1. Problema
Determinar si un estudiante puede postular a un taller.

## 2. Entrada
- edad
- promedio

## 3. Proceso
- Si edad >= 15 y promedio >= 14, puede postular.
- En caso contrario, no puede postular.

## 4. Salida
- Mostrar el resultado.

## 5. Código en Python

````python
edad = int(input("Edad: "))
promedio = float(input("Promedio: "))

if edad >= 15 and promedio >= 14:
    print("Puede postular")
else:
    print("No puede postular")
````
