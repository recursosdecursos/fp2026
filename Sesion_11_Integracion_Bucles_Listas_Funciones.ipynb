# Ejercicio 1: Registro de notas con funciónes

## Resolución guiada aplicada

1. Comprender: un registro de notas combina menú, lista y cálculo; las funciónes ayudan a ordenar el programa.
2. Entradas: opciónes de menú y notas.
3. Proceso: guardar notas en lista y usaráfunción para promedio.
4. Salida y pruebas: notas registradas y promedio. Prueba con: mostrar sin notas, registrar dos notas, calcular promedio y salir.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
def calcular_promedio(notas):
    suma = 0
    for nota in notas:
        suma = suma + nota
    return suma / len(notas)

notas = []
opción = ""

while opción != "4":
    print("1. Registrar nota")
    print("2. Mostrar notas")
    print("3. Calcular promedio")
    print("4. Salir")
    opción = input("Opción: ")

    if opción == "1":
        notas.append(float(input("Nota: ")))
    elif opción == "2":
        print("Notas:", notas)
    elif opción == "3":
        if len(notas) > 0:
            print("Promedio:", calcular_promedio(notas))
        else:
            print("No hay notas registradas")
    elif opción == "4":
        print("Fin")
    else:
        print("Opción inválida")
```

# Ejercicio 2: Lista de compras modular

## Resolución guiada aplicada

1. Comprender: una lista de compras necesita guardar productos y calcular total sin mezclar todo en un solo bl¿qué.
2. Entradas: productos y precios.
3. Proceso: guardar diccionarios en una lista y calcular total con función.
4. Salida y pruebas: productos y total. Prueba con: sin productos; luego dos productos.
5. Código: recién ahora revisa o escribe la celda de Python.


```python
def calcular_total(productos):
    total = 0
    for producto in productos:
        total = total + producto["precio"]
    return total

productos = []
opción = ""

while opción != "3":
    print("1. Agregar producto")
    print("2. Mostrar total")
    print("3. Salir")
    opción = input("Opción: ")

    if opción == "1":
        nombre = input("Producto: ")
        precio = float(input("Precio: "))
        productos.append({"nombre": nombre, "precio": precio})
    elif opción == "2":
        print("Productos:", productos)
        print("Total:", calcular_total(productos))
    elif opción == "3":
        print("Fin")
    else:
        print("Opción inválida")
```
