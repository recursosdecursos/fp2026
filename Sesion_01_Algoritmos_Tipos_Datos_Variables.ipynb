## Mini laboratorio: salidas, variables y tipos básicos


````python
nombre = "Ana"      # str
edad = 16           # int
promedio = 17.8     # float
aprobo = True       # bool

print("Nombre:", nombre)
print("Edad:", edad)
print("Promedio:", promedio)
print("¿Aprobó?:", aprobo)

print(type(nombre))
print(type(edad))
print(type(promedio))
print(type(aprobo))
````

## Mini laboratorio: `input()` siempre recibe texto


````python
edad_texto = input("Ingresa tu edad: ")
edad = int(edad_texto)

promedio_texto = input("Ingresa tu promedio: ")
promedio = float(promedio_texto)

print("El próximo año tendrás:", edad + 1)
print("Tu promedio es:", promedio)
````

# Ejercicio 2: Ficha escolar básica

## Resolución guiada aplicada

1. Comprender: necesitamos construir una ficha con datos reales de un estudiante, como si fuera un registro escolar.
2. Entradas: `nombre`, `edad`, `grado`, `seccion` y `promedio`.
3. Proceso: leer datos con `input()`, convertir edad con `int()` y promedio con `float()`.
4. Salida y pruebas: una ficha ordenada con etiquetas. Prueba con: edad 16 y promedio 17.5; luego prueba con otro estudiante.
5. Código: recién ahora revisa o escribe la celda de Python.


````python
nombre = input("Nombre: ")
edad = int(input("Edad: "))
grado = input("Grado: ")
seccion = input("Sección: ")
promedio = float(input("Promedio: "))

print("===== Ficha escolar =====")
print("Nombre:", nombre)
print("Edad:", edad)
print("Grado:", grado)
print("Sección:", seccion)
print("Promedio:", promedio)
````
