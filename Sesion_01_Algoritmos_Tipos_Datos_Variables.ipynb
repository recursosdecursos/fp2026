## Mini laboratorio de variables y tipos de datos

Una variable guarda un dato. En esta sesión usaremos cuatro tipos básicos:

- Texto: `str`
- Entero: `int`
- Decimal: `float`
- Lógico: `bool`

````python
nombre = "Ana"          # texto / str
edad = 16               # entero / int
promedio = 17.8         # decimal / float
aprobo = True           # lógico / bool

print("Nombre:", nombre)
print("Edad:", edad)
print("Promedio:", promedio)
print("¿Aprobó?:", aprobo)
````

## Mini laboratorio: revisar el tipo de una variable

Python puede decirnos el tipo de dato usando `type()`.

````python
print(type(nombre))
print(type(edad))
print(type(promedio))
print(type(aprobo))
````

## Mini laboratorio: `input()` siempre recibe texto

Aunque escribas un número, `input()` lo recibe como texto. Para hacer cálculos se convierte con `int()` o `float()`.

````python
edad_texto = input("Ingresa tu edad: ")
edad = int(edad_texto)

nota_texto = input("Ingresa tu promedio: ")
promedio = float(nota_texto)

print("El próximo año tendrás:", edad + 1)
print("Tu promedio es:", promedio)
````

# Ejercicio 3: Ficha escolar básica
